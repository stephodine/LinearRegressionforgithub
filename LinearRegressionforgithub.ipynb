```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("Salary Data.csv")
```

```python
#to check for errors
dataset.isnull().sum()
```

```python
#to remove errors and emoty spaces
dataset = dataset.dropna()
```

```python
#this displays all the rows
pd.set_option("display.max_rows",None)
```

```python
dataset
```

```python
#independent variable=x
#dependent=y
#simple linear regression 

#setting the dependent and independent variable to the specified columns
x=dataset["Years of Experience"]
y=dataset["Salary"]
```

```python
#spliting the datasets in two for training and testing
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y, test_size=0.2,random_state=0)
```

```python
len(x_train)
```

```python
#fitting the model/testing the model
from sklearn.linear_model import LinearRegression
```

```python
#using the LinearRegression model
model = LinearRegression()
```

```python
#to check the shape and ensuring it is a two dimensional array if its not then  convert to 2D
import numpy as np
x_train.shape
```

```python
#converting to 2D
#when we use one column as independent variable then we need tp convert to 2D
x_train = np.array(x_train).reshape(298,1)
x_train.shape
```

```python
len(x_test)
```

```python
#converting the x_test to a 2D
x_test = np.array(x_test).reshape(75,1)
x_test.shape
```

```python
#fitting the variables
model.fit(x_train,y_train)
```

```python
#after fitting the model, we can get the m and the c
c = model.intercept_
```

```python
m = model.coef_
```

```python
#formula is y =  mx + c
#where y is the dependent variable
#x is the independent variable
#c is the intercept of the line
#m is the coefficient
```

```python
#substituting into the equation
y_pred= m*x_train + c
```

```python
y_pred

```
